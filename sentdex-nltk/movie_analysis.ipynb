```python
import nltk
import random
import string
from nltk.corpus import stopwords 
```

```python
from nltk.corpus import movie_reviews
```

```python
doc=[]
for cat in movie_reviews.categories():
    for fileid in movie_reviews.fileids(cat):
        doc.append((list(movie_reviews.words(fileid)),cat))
```

```python
# print(doc[0])
# to shuffle docs for testing data
# random.shuffle(doc)
```

```python
stop_words=set(stopwords.words('english'))
```

```python
punc=[]
for p in string.punctuation:
    punc.append(p)
```

```python
all_words=[]
for w in movie_reviews.words():
    w=w.lower().encode('utf8')
    if((w not in punc) and (w not in stop_words)):
        all_words.append(w)
```

```python
new_words=nltk.FreqDist(all_words)
```

```python
new_words.most_common()
```

```python
new_words['bad']
```

```python
# finding top 1000 words in document
def find_features(document):
    words=set(document)
    features={}
    for w in list(new_words.keys())[:1000]:
#         if (w in words) and len(w)>1:
        features[w]={w in words}
    return features    
```

```python
feature_Set=[(find_features(rev[0]),cat) for (rev,cat) in doc]    
# find_features(doc[0][0])
```
