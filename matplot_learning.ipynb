```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x=[1,2,3]
y=[2,3,1]
```

```python
plt.plot(x,y,'r')
# plt.ylabel('M label')
# plt.xlabel('A label')
plt.title('MAYANK')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r',label="B")
plt.subplot(1,2,2)
plt.plot(y,x,'b',label="A")
plt.legend()
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.9,0.1,0.1,0.1])
axes2=fig.add_axes([0.0,0.0,0.8,0.8])
axes2.plot(x,y,label="First",color='green',marker='o')
axes2.plot(x,[2,4,8],label="Second",color="red",marker='o',alpha=0.5)
axes2.legend()
```

```python
fig,axes=plt.subplots(1,2)
axes[1].plot(x,y)
# plt.tight_layout()
fig.axes[0].plot(x,y)
fig.tight_layout()
```
